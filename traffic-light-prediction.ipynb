```python
def label(s):
    if s == 'stop' :
        return 0
    return 1

train_images = []
train_labels = []
test_images = []
test_labels = []

for index,detail in day1.iterrows():
    if index%2==1:
        continue
    im=cv2.imread("/kaggle/input/lisa-traffic-light-dataset/daySequence1/daySequence1/frames/"+detail['Filename'][8:])
    im = im[detail['Upper left corner Y']:detail['Lower right corner Y'],detail['Upper left corner X']:detail['Lower right corner X'],]  
    im=cv2.resize(im,(224,224))
    im = im.astype(np.float64) / 255          
    
    train_images.append(np.array(im))
    train_labels.append(label(detail['Annotation tag']))
```

```python
for i in range(len(train_images)):
    if i%4==0:
        test_images.append(train_images[i])
        test_labels.append(train_labels[i])
t_images = [train_images[i] for i in range(len(train_images)) if i%4!=0]
t_labels = [train_labels[i] for i in range(len(train_labels)) if i%4!=0] 
train_images=t_images
train_labels=t_labels
print(len(train_images))
print(len(test_images))
```

```python
class AlexNet(nn.Module):
    def __init__(self, num_classes=10):
        super(AlexNet, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 96, kernel_size=11, stride=4, padding=2),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),
            nn.Conv2d(96, 256, kernel_size=5, padding=2),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),
            nn.Conv2d(256, 384, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(384, 384, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(384, 384, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),
        )
        self.avgpool = nn.AdaptiveAvgPool2d((6, 6))
        self.classifier = nn.Sequential(
            nn.Dropout(),
            nn.Linear(384*6*6, 4096),
            nn.ReLU(inplace=True),
            nn.Dropout(),
            nn.Linear(4096, 4096),
            nn.ReLU(inplace=True),
            nn.Linear(4096, num_classes),
        )

    def forward(self, x):
        x = self.features(x)
        x = self.avgpool(x)
        x = torch.flatten(x, 1)
        x = x.view(-1,384*6*6 ) 
        x = self.classifier(x)
        return x
```

```python
if torch.cuda.is_available():
    device = torch.device("cuda")  
else:
    device = torch.device("cpu") 

model = AlexNet(num_classes=2).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

for epoch in range(1): 
    running_loss = 0.0
    for i in range(len(train_images)):
        inputs, labels = train_images[i],train_labels[i]
        im=torch.zeros(size=(3,224,224))
        for j in range(224):
            for z in range(224):
                im[0][j][z]=inputs[j][z][0]
                im[1][j][z]=inputs[j][z][1]
                im[2][j][z]=inputs[j][z][2]
        optimizer.zero_grad()
        im=im.to(device)
        outputs = model(im)
        label=torch.tensor(labels,dtype=torch.int64).long()
        if label==1:
            label=torch.tensor([[0,1]]).float()
        else:
            label=torch.tensor([[1,0]]).float()
        label=label.to(device)
        outputs=outputs.to(device)
        loss = criterion(outputs, label)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        if i % 50 == 49:  # print every 50 mini-batches
            print('[%d, %5d] loss: %.10f' %
                  (epoch + 1, i + 1, running_loss / 2000))
            running_loss = 0.0
        if i>1000:
            break
```
